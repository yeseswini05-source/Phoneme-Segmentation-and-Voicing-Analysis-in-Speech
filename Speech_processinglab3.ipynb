```python
import librosa
import matplotlib.pyplot as plt
import numpy as np

signal, sr = librosa.load("sentence1.m4a", sr=None)
signal2, sr2 = librosa.load("sentence2.m4a", sr=None)


time = np.linspace(0, len(signal)/sr, len(signal))

plt.figure(figsize=(14,4))
plt.plot(time, signal)
plt.title("Sentence 1 Waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
def plot_segment(start, end, label):
    seg = signal[int(start*sr):int(end*sr)]
    t = np.linspace(0, len(seg)/sr, len(seg))

    plt.figure(figsize=(5,2))
    plt.plot(t, seg)
    plt.title(label)
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.show()
```

```python
plot_segment(0.2,0.4,"Unvoiced (sh)")
plot_segment(3.3,3.6,"Voiced (b)")
```

```python
time2 = np.linspace(0, len(signal2)/sr2, len(signal2))

plt.figure(figsize=(14,4))
plt.plot(time2, signal2)
plt.title("Sentence 2 Waveform")
plt.show()
```

```python
def plot2(start,end,label):
    seg = signal2[int(start*sr2):int(end*sr2)]
    plt.figure(figsize=(5,2))
    plt.plot(seg)
    plt.title(label)
    plt.show()

plot2(0.1,0.3,"Fricative (s)")
plot2(0.6,0.9,"Approximant (y)")
```
